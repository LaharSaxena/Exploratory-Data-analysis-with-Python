```python
#Exploring fields related to rating...
df_rating = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
df_rating
```

Inference:
1. Maximum number of people has not rated the zomato
2. When Rating is between 4.5 to 4.9---> Excellent
3. When Rating is between 4.0 to 3.4---> Very Good
4. when Rating is between 3.5 to 3.9---> Good
5. when Rating is between 3.0 to 3.4---> Average
6. when Rating is between 2.5 to 2.9---> Average
7. when Rating is between 2.0 to 2.4---> Poor
